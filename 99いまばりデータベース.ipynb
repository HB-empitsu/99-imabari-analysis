```python
import sqlite3
```

```python
# 設定

DATABASE_FILE = "imabariHosp.db"
TABLE_NAME = "hospital"
```

# 今日のみ取得

```python
# データベース準備
conn = sqlite3.connect(DATABASE_FILE)
c = conn.cursor()
```

```python
# 今日のデータを抽出する
c.execute("SELECT DISTINCT date, week FROM hospital WHERE date = date('now') ORDER BY date ASC")
```

```python
dates = c.fetchall()
dates
```

```python
posts_by_date = {}

for date in dates:
    c.execute("SELECT * FROM hospital WHERE date=?",(date[0],))
    posts_by_date[date[0]] = c.fetchall()
    print(posts_by_date[date[0]])
```
