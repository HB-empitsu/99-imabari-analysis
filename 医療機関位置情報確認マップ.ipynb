```python
import pandas as pd
```

# スプレッドシート読み込み

```python
# 医療機関の位置情報
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQijVNaEWw2giRgQJSaBsJJAzdzhR47dQ3NU1eBj9lSCv8-bZnUjD6e2CFUV_YqQqs-xsdKBAWM8LOb/pub?gid=0&single=true&output=csv"
```

```python
# CSVを読み込み
df = pd.read_csv(url)
```

```python
# 内容確認
df
```

# 地図表示

```python
import folium
```

```python
# 国土地理院の地図を利用
# https://maps.gsi.go.jp/development/ichiran.html

map = folium.Map(
    # 中心は今治市役所
    location=[34.0662403, 132.9976865],
    zoom_start=14,
    tiles="https://cyberjapandata.gsi.go.jp/xyz/pale/{z}/{x}/{y}.png",
    attr='&copy; <a href="https://maps.gsi.go.jp/development/ichiran.html">国土地理院</a>',
)
```

```python
for i, r in df.iterrows():

    folium.Marker(
        # 緯度・経度
        location=[r.lat, r.lon],
        # 吹き出しに医療機関名
        popup=folium.Popup(f'<p>{r["name"]}</p>', max_width=300),
        # アイコン（赤）
        icon=folium.Icon(color="red")
    ).add_to(map)
```
